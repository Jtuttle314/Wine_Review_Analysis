# Dataset Details

## The data consists of 10 fields:

- #### Country: The country that the wine is from
- #### Description: Few sentences from a sommelier describing the wine's taste, smell, look, feel, etc.
- #### Designation: The vineyard within the winery where the grapes that made the wine are from
- #### Points: The number of points WineEnthusiast rated the wine on a scale of 1-100 (though they say they only post reviews for wines that score >=80)
- #### Price: The cost for a bottle of the wine
- #### Province: The province or state that the wine is from
- #### Region 1: The wine growing area in a province or state (ie Napa)
- #### Region 2: Sometimes there are more specific regions specified within a wine growing area (ie Rutherford inside the Napa Valley), but this value can sometimes be blank
- #### Variety: the type of grapes used to make the wine (ie Pinot Noir)
- #### Winery: The winery that made the wine

### Data Types of the columns

- country    :    object
- description:    object
- designation:    object
- points     :    int64
- price      :    float64
- province   :    object
- region_1   :    object
- region_2   :    object
- variety    :    object
- winery     :    object

## Prediction using a Decision Tree

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=10)
```

#### Commented below section as removing the Nan values does not give improvement in prediction. Roughly a 0.5 percent decrease  

```python
# print("dataset size before removing records with Nan's:",len(parsed_data))
# parsed_data = parsed_data.dropna(subset=['province','variety','country','price','points','region_1'], how='any')
# print("dataset size after removing records with Nan's:",len(parsed_data))
```

```python
## create the x and y columns
x_unsplit = wine[['variety', 'province', 'price', 'winery', 'region_1']]
y_unsplit = wine['points']
## one hot encode the data
x_unsplit = pd.get_dummies(x_unsplit, columns=['variety', 'province', 'winery', 'region_1'])
print(x_unsplit.shape, y_unsplit.shape)
## get the training and testing data
X_train, X_test, y_train, y_test = train_test_split(x_unsplit, y_unsplit, random_state=1, train_size=0.65)
```

#### Fill the null values in price column with the mean price 

```python
X_train_35 = X_train.fillna({"price": x_unsplit['price'].mean()})
X_test_35 = X_test.fillna({"price":x_unsplit['price'].mean()})
clf.fit(X_train_35, y_train)
```

```python
y_predictions = clf.predict(X_test_35)
```

```python
from sklearn.metrics import accuracy_score
dt_acc = accuracy_score(y_test,y_predictions)
print(dt_acc)
```

## Accuracy = 69.218% with features having Nan values
## Accuracy = 68.787% with features not having Nan values
